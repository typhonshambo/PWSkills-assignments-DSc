## Q3

The subplot() function in Matplotlib is used to create multiple plots (subplots) within the same figure.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for line plots
x = np.array([0, 1, 2, 3, 4, 5])

# Line 1
y1 = np.array([0, 100, 200, 300, 400, 500])

# Line 2
y2 = np.array([50, 20, 40, 20, 60, 70])

# Line 3
y3 = np.array([10, 20, 30, 40, 50, 60])

# Line 4
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create subplots
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title('Line 1')

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('Line 2')

plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title('Line 3')

plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('Line 4')

# Adjust layout for better visualization
plt.tight_layout()

# Show the plots
plt.show()
```

## Q4

A bar plot (or bar chart) is a graphical representation of data where individual bars represent different categories, and the lengths of the bars correspond to the values they represent

```python
import matplotlib.pyplot as plt
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Bar Plot
plt.figure(figsize=(8, 5))
plt.bar(company, profit, color='skyblue')
plt.xlabel('Company')
plt.ylabel('Profit (in millions)')
plt.title('Company Profits (Bar Plot)')
plt.show()

# Horizontal Bar Plot
plt.figure(figsize=(8, 5))
plt.barh(company, profit, color='lightcoral')
plt.xlabel('Profit (in millions)')
plt.ylabel('Company')
plt.title('Company Profits (Horizontal Bar Plot)')
plt.show()
```

## Q5

A box plot (box-and-whisker plot) is a statistical visualization that displays the distribution of a dataset based on a five-number summary: minimum, first quartile (Q1), median, third quartile (Q3), and maximum. 

```python
import matplotlib.pyplot as plt
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Combine data into a single list
data = [box1, box2]

# Create a box plot
plt.figure(figsize=(8, 5))
plt.boxplot(data, labels=['Box 1', 'Box 2'])
plt.xlabel('Box')
plt.ylabel('Values')
plt.title('Box Plot of box1 and box2')
plt.show()
```
