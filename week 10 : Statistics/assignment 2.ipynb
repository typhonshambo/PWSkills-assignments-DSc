# Statistics Assignment

## Q1. Measures of Central Tendency

The three measures of central tendency are:
1. Mean
2. Median
3. Mode

## Q2. Difference between Mean, Median, and Mode

Mean is the average, median is the middle value, and mode is the most frequent value. They are used to measure the central tendency of a dataset, providing insights into its typical or central values.

## Q5. Measures of Dispersion

Range, Variance, and Standard Deviation are measures of dispersion used to describe the spread of a dataset. For example, a high standard deviation indicates greater variability in the data.

## Q6. Venn Diagram

A Venn diagram is a visual representation of the relationships between sets. It uses overlapping circles to show commonalities and differences between sets.

## Q7. Set Operations

## Q8. Skewness

Skewness in data indicates asymmetry. A right-skewed distribution has a longer right tail.

## Q9. Median in Right Skewed Data

In right-skewed data, the median is typically less than the mean.

## Q10. Covariance vs Correlation

Covariance measures the degree of joint variability between two random variables, while correlation standardizes this measurement, making it easier to interpret.

## Q11. Sample Mean Formula

Sample Mean Formula: \( \bar{X} = \frac{1}{n} \sum_{i=1}^{n} X_i \)
Example Calculation: \( \bar{X} = \frac{2+4+6+8+10}{5} = 6 \)

## Q12. Relationship in Normal Distribution

In a normal distribution, the mean, median, and mode are approximately equal.

## Q13. Covariance vs Correlation

Covariance measures the direction of the linear relationship between two variables, while correlation additionally measures the strength and standardizes the result.

## Q14. Outliers Effect

Outliers can significantly affect the mean and standard deviation, making them less representative of the central tendency and dispersion of the majority of the data.
